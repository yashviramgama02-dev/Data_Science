```python
class Vehicle:
    def general_usage(self):
        print("general usage : transporation")
        
class car(Vehicle):
    def __init__(self):
        print("i am car")
        self.wheels = 4
        self.has_roof = True
        print("wheels",self.wheels)
        
    def specific_usage(self):
        self.general_usage()
        print("specific usage: trip with family, commute to work")      
          
class Motorcycle(Vehicle):
    
    def __init__(self):
        print("i am motor cycle")
        self.wheels = 2
        self.has_roof = False
        print("wheels",self.wheels)
        
    def specific_usage(self):
        self.general_usage()
        print("specific usage: road trip, racing")
        
c = car()
m = Motorcycle()
m.specific_usage()

        
    
```

```python
class Animal:
    def __init__(self, habitat):
        self.habitat = habitat

    def print_habitat(self):
        print(self.habitat)

    def sound(self):
        print("Some Animal Sound")


class Dog(Animal):
    def __init__(self):
        super().__init__("Kennel")

    def sound(self):
        print("Woof woof!")


x = Dog()
x.print_habitat()
x.sound()
```

```python
class Animal:
    def sound(self):
        print("Animal sound")

class Dog(Animal):
    def sound(self):
        super().sound()
        print("Bark")
        
d = Dog()
d.sound()


```
