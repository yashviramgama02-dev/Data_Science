```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('movie_revenues.csv')
df.head(3)
```

```python
df.revenue.describe()
```

```python
df['revenue_mln'] = df['revenue'].apply(lambda x: x / 1_000_000)
df.revenue_mln.describe()
```

```python
_, mean, std, *_ = df.revenue_mln.describe()
```

```python
(mean,std)
```

OUTLIERS DETECTION USING SIMPLE Z SCORE

```python
def z_score(x,mean,std):
    return (x - mean) / std
```

```python
df['z_score'] = df.revenue_mln.apply(lambda x: z_score(x,mean,std))
df.head(3)
```

```python
df[df.z_score>3]
```

Outlier detection using modified Z score

```python
def get_mad(s):
    median = np.median(s)
    diff = abs(s - median)
    mad = np.median(diff)
    return mad
```

```python
MAD = get_mad(df.revenue_mln)
median = np.median(df.revenue_mln)
MAD, median
```

```python
def get_modified_z_score(x, median, MAD):
    return 0.6745*(x-median)/MAD
```

```python
get_modified_z_score(2711, median, MAD)
```

```python
df['mod_z_score'] = df.revenue_mln.apply(lambda x: get_modified_z_score(x, median, MAD))
df.head(3)
```

```python
df[df.mod_z_score>3.5]
```
