SIGMOID

```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))
   
```

```python
sigmoid(100)
```

TAN H

```python
def tanh(x):
  return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(-560)
```

```python
tanh(4)
```

```python
tanh(1)
```

RELU

```python
def relu(x):
    return max(0,x)
```

```python
relu(-100)
```

```python
relu(8)
```

Leaky ReLU

```python
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(8)
```

```python
leaky_relu(-100)
```
